```python
import pandas as pd
```

```python
res = []
with open("netflix/movie_titles.csv", encoding="ISO-8859-1") as file:
    lines = file.readlines()
    for line in lines:
        data = line.split("\n")[0].split(",",2)
        res.append(data[1:])
        
```

```python
df = pd.DataFrame(res,columns=["year", "title"])
df
```

```python
df.to_csv("movie_titles_new.csv", index=False)
```
